```python
import pandas as pd
```

```python
# using column by column

hey_arnold = pd.DataFrame ({
    "Character_in_show" : ["Arnold","Gerald","Helga","Phoebe","Harold","Eugene"],
    "Football_Shaped_Head" : [True, False, False, False, False, False],
    "Height" : ["average","tallish","tallish","short","tall","short"],
    "color_of_hair":["blond","black","blond","black","unknown","red"]
    })

hey_arnold
```

```python
# using row by row

hey_arnold_row = pd.DataFrame([
    {"Character_in_show":"Arnold","Football_Shaped_Head":True,"Height":"average","color_of_hair":"blond"},
    {"Character_in_show":"Gerald","Football_Shaped_Head":False,"Height":"tallish","color_of_hair":"black"},
    {"Character_in_show":"Helga","Football_Shaped_Head":False,"Height":"tallish","color_of_hair":"blond"},
    ])
hey_arnold_row
```

```python
# Rename columns using .rename()
# Height column was not renamed because it didnt change

hey_arnold_renamed = hey_arnold.rename(columns = {"Character_in_show":"Character",
                                                 "color_of_hair":"Hair Color",
                                                  "Football_Shaped_Head":"Football Head"
                                                 })
hey_arnold_renamed
```

```python
# Rearrange columns using [[]]

hey_arnold_alpha = hey_arnold_renamed[["Character","Height","Hair Color","Football Head"]]
hey_arnold_alpha
```
