```python
import pandas as pd
import numpy as np
csv_path = "Week_04_Pandas_CSV_XLSX/Soccer2018Data.csv"
soccer_2018_df = pd.read_csv(csv_path, low_memory=False)
soccer_2018_df.head()
```

```python
# Collect a list of all the unique values in "Preferred Position" using .unique()
soccer_2018_df["Preferred Position"].unique()
```

```python
soccer_unique = soccer_2018_df["Preferred Position"].unique()
print(soccer_unique)
```

```python
# Looking only at strikers (ST) to start in "Preferred Position"

strikers_2018_df = soccer_2018_df.loc[soccer_2018_df["Preferred Position"] == "ST", :]
strikers_2018_df.head()
```

```python
# Sort the DataFrame by the values in the "ST" column to find the worst 
# Reset the index so that the index is now based on the sorting locations using .reset_index(drop = True)

strikers_2018_df = strikers_2018_df.sort_values("ST")
strikers_2018_df = strikers_2018_df.reset_index(drop=True)
strikers_2018_df.head()
```

```python
# Save all of the information collected on the worst striker

worse_striker = strikers_2018_df.loc[0, :]
worse_striker
```
