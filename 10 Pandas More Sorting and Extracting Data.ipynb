```python
import pandas as pd
csv_path = "Week_04_Pandas_CSV_XLSX/ufoSightings.csv"
ufo_df = pd.read_csv(csv_path)
ufo_df.head()
```

```python
# Remove rows with missing data

clean_ufo_df = ufo_df.dropna(how="any")
clean_ufo_df.count()
```

```python
clean_ufo_df.dtypes
```

```python
# Convert "duration (seconds)" column's values to numeric

clean_ufo_df["duration (seconds)"] = pd.to_numeric(clean_ufo_df["duration (seconds)"])
clean_ufo_df["duration (seconds)"]
```

```python
type(clean_ufo_df["duration (seconds)"])
```

```python
clean_ufo_df["duration (seconds)"].dtypes
```

```python
# Now that "duration (seconds)" is a numeric time, it can now be added using .sum()
#total_ufo_duration = clean_ufo_df.sum()
#total_ufo_duration
```

```python
# Filter the data so only those sightings in the US are in the DataFrame

usa_ufo_df = clean_ufo_df.loc[clean_ufo_df["country"] == "us", :]
usa_ufo_df
```

```python
# Count how many sightings have occured within each state using .values_counts()

state_counts = usa_ufo_df["state"].value_counts()
state_counts.head()
```

```python
# Use Groupby in order to separate the data into fields according to "state" using df.groupby([])

grouped_usa_df = usa_ufo_df.groupby(["state"])
```

```python
# Object returned in a "GroupBy" object cannot be viewed normally

print(grouped_usa_df)
```

```python
usa_ufo_df
```

```python
# Data function such as df.count().head() has to be used to visualize "GroupBy" object

grouped_usa_df.count().head(10)
```

```python
# Since "duration (seconds)" was converted to a numeric time, it can now be summed up per state

state_duration = grouped_usa_df["duration (seconds)"].sum()
state_duration.head()
```

```python
# Create a new DataFrame using both duration and count

state_summary_table = pd.DataFrame({"Number of Sightings" :  state_counts, 
                                    "Total Visit Time" : state_duration})
state_summary_table.head()
```
