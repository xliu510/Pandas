```python
import pandas as pd
import numpy as np
pokemon = "Week_04_Pandas_CSV_XLSX/Pokemon.csv"
pokemon_pd = pd.read_csv(pokemon)
pokemon_pd.head()
```

```python
# Create a new pokemon table

pokemon_type = pokemon_pd.loc[:,["Type 1","HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]]
pokemon_type # 800 row by 7 column
```

```python
# Use a lambda function to capitalizer all values in column 'Type 1'
capitalizer = lambda x : x.upper()
pokemon_type['Type 1'] = pokemon_type['Type 1'].apply(capitalizer)
pokemon_type
```

```python
# Find Attack that is >= 50 with a function
def attack_above50(x):
    if x >= 50:
        return True
    else:
        return False
```

```python
pokemon_type.Attack = pokemon_type.Attack.apply(attack_above50)
```

```python
pokemon_type
```

```python
# Create the GroupBy object based on the "Type 1" column using .groupby([])

pokemon_group = pokemon_type.groupby(["Type 1"])
pokemon_group.head() # 89 row by 7 column
```

```python
# Calculate averages for combat stats using the .mean() method

pokemon_comparison = pokemon_group.mean()
pokemon_comparison
```

```python
# Add the all the value of each row

pokemon_comparison["Total"] = pokemon_comparison["HP"] + pokemon_comparison["Attack"] + \
                            pokemon_comparison["Defense"] + pokemon_comparison["Sp. Atk"] + \
                            pokemon_comparison["Sp. Def"] + pokemon_comparison["Speed"]

pokemon_comparison["Total"]
```

```python
pokemon_comparison
```

```python
# Sort to find the strongest pokemon by "Total" column using .sort_values

strongest_pokemon = pokemon_comparison.sort_values(["Total"], ascending=False)
strongest_pokemon
```

```python
# Reset indexing using .reset_index(inplace=True)

strongest_pokemon.reset_index(inplace=True)
strongest_pokemon
```

```python
# Save output to Excel using .to_excel("", index = False)

strongest_pokemon.to_excel("Week_04_Pandas_CSV_XLSX/pokemon_rankings.xlsx", index = False)
```
